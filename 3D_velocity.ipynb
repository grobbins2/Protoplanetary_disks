```python
import readfargo3d as fg
import numpy as pl
import numpy as np
import matplotlib.pyplot as plt

path3d = './p3disof_test01jup/'
path2d = './bae2017_01jup1400x512/'

itime = 100
r0 = 1.0
fargo3d    = fg.frame(itime,rho=True,vphi=True,vr=True,vtheta=True,dir=path3d)
fargo3d_init    = fg.frame(0,rho=True,vphi=True,vr=True,vtheta=True,dir=path3d)
fargo2d    = fg.frame(itime,rho=True,vphi=True,vr=True,vtheta=False,dir=path2d)
fargo2d_init    = fg.frame(0,rho=True,vphi=True,vr=True,vtheta=False,dir=path2d)
print('done')
```

```python
#vphi versus r for varying scale heights(0H,1H,3H) and 2d/3d comparison
#one scale height, H/R = 0.05, indice of +11 each time
fig, ax = plt.subplots()
ax.plot(fargo2d.r, (fargo2d.vphi[:,:].mean(axis=0)-fargo2d_init.vphi[:,:].mean(axis=0))/fargo2d_init.vphi[:,:].mean(axis=0),linewidth=1., label = '2D')
ax.plot(fargo3d.r, (fargo3d.vphi[:,:,33].mean(axis=0)-fargo3d_init.vphi[:,:,33].mean(axis=0))/fargo3d_init.vphi[:,:,33].mean(axis=0),linewidth=1., label = '0H')
ax.plot(fargo3d.r, (fargo3d.vphi[:,:,44].mean(axis=0)-fargo3d_init.vphi[:,:,44].mean(axis=0))/fargo3d_init.vphi[:,:,44].mean(axis=0),linewidth=1., label = '1H')
ax.plot(fargo3d.r, (fargo3d.vphi[:,:,65].mean(axis=0)-fargo3d_init.vphi[:,:,65].mean(axis=0))/fargo3d_init.vphi[:,:,65].mean(axis=0),linewidth=1., label = '3H')

ax.set_xlabel('r')
ax.set_ylabel('vphi')
ax.set_title('Vphi vs. r, 3D-2D comparison at 0H,1H,3H, azimuthally avg')
plt.legend()
plt.savefig('vphi_r_scale2D3D_'+str(itime*10)+'_orbits',dpi=300,bbox_inches='tight')
```

```python
#vrad versus r for varying scale heights(0H,1H,3H) and 2d/3d comparison
fig, ax = plt.subplots()
ax.plot(fargo3d.r, fargo3d.vtheta[:,:,33].mean(axis=0),linewidth=1., label = '0H')
ax.plot(fargo3d.r, (fargo3d.vtheta[:,:,44]).mean(axis=0),linewidth=1., label = '1H')
ax.plot(fargo3d.r, fargo3d.vtheta[:,:,65].mean(axis=0),linewidth=1., label = '3H')
ax.set_xlabel('r')
ax.set_ylabel('vtheta')
ax.set_title('vtheta vs. r, 3D comparison at 0H,1H,3H, azimuthally avg')
plt.legend()
plt.savefig('vtheta_r_scale3D01_'+str(itime*10)+'_orbits',dpi=300,bbox_inches='tight')
```

```python
#vrad versus r for varying scale heights(0H,1H,3H) and 2d/3d comparison
fig, ax = plt.subplots()
ax.plot(fargo2d.r, fargo2d.vr[:,:].mean(axis=0),linewidth=1., label = '2D')
ax.plot(fargo3d.r, fargo3d.vr[:,:,33].mean(axis=0),linewidth=1., label = '0H')
ax.plot(fargo3d.r, fargo3d.vr[:,:,44].mean(axis=0),linewidth=1., label = '1H')
ax.plot(fargo3d.r, fargo3d.vr[:,:,65].mean(axis=0),linewidth=1., label = '3H')

ax.set_xlabel('r')
ax.set_ylabel('vrad')
ax.set_title('vrad vs. r, 3D-2D comparison at 0H,1H,3H, azimuthally avg')
plt.legend()
plt.savefig('vrad_r_scale2D3D_'+str(itime*10)+'_orbits',dpi=300,bbox_inches='tight')
```
